```python
match['date'] = pd.to_datetime(match['date'])
```

```python
# Creating predictor for machine learning
```

```python
# cree une colonne 'venue_code' qui met 1 si l equipe est home sur VENU et 0 si home sur VENU
match["venue_code"] = match["venue"].astype("category").cat.codes
match
```

```python
match['opp_code'] = match['opponent'].astype('category').cat.codes

# it will get the first part of hours on TIME coloumn , ex :
# 16:30 will be just 16
match['hour'] = match['time'].str.replace(':.+',"",regex=True).astype('int')
```

```python
#so now , I'll make each day of the week to number int 
match['day_code'] = match['date'].dt.dayofweek
```

```python
match
```

```python
match['target']=(match['result']=='W').astype('int')
```

```python
#creating our inital maching learning model 
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf  = RandomForestClassifier(n_estimators=50,min_samples_split=10,random_state=1)
```

```python
# donnes d entrainement tout les row inferieur a 1er janvire 2022
train  = match[match['date'] <'2022-01-01']

#donnes de test tout les row superieur a 01 janvier 2022
test = match[match['date'] >='2022-01-01']
```

```python
#les colonnes transformer en nombre modelable seront les colonnes de prediction 
predictors = ["venue_code","opp_code","hour","day_code"]
```

```python
rf.fit(train[predictors],train["target"])
```

```python
preds  = rf.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(test["target"],preds)
```

```python
combined  = pd.DataFrame(dict(actual=test["target"],prediction=preds))
```

```python
from sklearn.metrics import precision_score 
```

```python
precision_score(test["target"],preds)
```

```python
group_matches  = match.groupby("team")
```

```python
def rolling_average(group , cols , new_cols):
    group  = group.sort_values('date')
    rolling_stats = group[cols].rolling(3,closed='left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group
    
```

```python
cols  = ['gf','ga','sh','sot','dist','fk','pk','pkatt']
new_cols  = [f"{c}_rolling" for c in cols]
```

```python
new_cols
```

```python
rolling_average(group , cols , new_cols)
```

```python
matches_rolling = match.groupby("team").apply(lambda x:rolling_average(x , cols , new_cols))
```

```python
matches_rolling
```

```python
def make_prediction(data , predictors):
    train  = data[data["date"]< '2022-01-01']
    test = data[data["date"]>='2022-01-01']
    rf.fit(train[predictors],train["target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["target"],predicted=preds),index=test.index)
    precision  = precision_score(test["target"],preds)
    return combined , precision
```

```python
combined , precision  = make_prediction(matches_rolling , predictors + new_cols)
```

```python
combined  = combined.merge(matches_rolling[['date','team','opponent','result']],left_index=True,right_index=True)
```
